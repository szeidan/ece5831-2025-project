# Module 3 - Logic Gate Implementation

### Introduction
#### In this notebook, we will test a LogicGate class that uses NumPy.
#### NumPy allows us to perform operations on entire lists (arrays) of numbers at once,
#### so we can apply AND, OR, XOR, etc. to multiple values without writing loops.

````python
from logic_gate import LogicGate
import numpy as np
````

## Create an instance of LogicGate so we can call its functions

````python
g = LogicGate()
print('LogicGate loaded.')
````

### Define Inputs
#### We will define two arrays a and b.
#### Each value is either 1 (True) or 0 (False).

````python
a = np.array([1, 0, 1, 0])
b = np.array([1, 1, 0, 0])
print("Input a:", a)
print("Input b:", b)
````

## AND Gate
# The AND gate outputs 1 only if both inputs are 1.

````python
print("AND Gate :", g.and_gate(a, b))
````

## NAND Gate
# The NAND gate is the opposite of AND.
# It outputs 0 only if both inputs are 1.

````python
print("NAND Gate:", g.nand_gate(a, b))
````

### OR Gate
#### The OR gate outputs 1 if at least one of the inputs is 1.

````python
print("OR Gate  :", g.or_gate(a, b))
````

### NOR Gate
#### The NOR gate is the opposite of OR.
#### It outputs 1 only if both inputs are 0.

````python
print("NOR Gate :", g.nor_gate(a, b))
````

### XOR Gate
#### The XOR gate outputs 1 only if the inputs are different.

````python
print("XOR Gate :", g.xor_gate(a, b))
````

# Summary
#### I have tested AND, NAND, OR, NOR, and XOR gates.
#### These operations work on entire NumPy arrays at once,
#### so we can quickly calculate logic for many values at the same time.
