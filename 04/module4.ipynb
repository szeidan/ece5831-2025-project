## Multilayer Perceptron
### Step and Sigmoid Functions
Letâ€™s implement a step function and a sigmoid function using NumPy.

```python
import numpy as np
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int32)
```

```python

x = np.array([-1.0, -2.0, 1.0, 1.5])
y = step_function(x)
print(y)  # Output: [0 0 1 1]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example input
x = np.array([-1.0, -2.0, 1.0, 1.5])
y = sigmoid(x)
print(y)  # Output: [0.26894142 0.11920292 0.73105858 0.81757448]
```

Plot Both Functions Using Matplotlib

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y_step = step_function(x)
y_sigmoid = sigmoid(x)

plt.plot(x, y_step, label='Step Function')
plt.plot(x, y_sigmoid, label='Sigmoid Function')
plt.title('Step vs Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

## Three Layer-MultiLayerPerceptron class
Let's implement three later-MLP class. Two inputs, 3 nodes in the Layer 1, 2nodes in L2, 2 output nodes

```python
class ThreeLayerMLP:
    def __init__(self):
        # Initialize weights and biases for each layer
        self.w1 = np.array([[1, 3, 5], [2, 4, 6]])
        self.b1 = np.array([0.1, 0.2, 0.3])

        self.w2 = np.array([[1, 4], [2, 5], [3, 6]])
        self.b2 = np.array([0.4, 0.3])

        self.w3 = np.array([[1, 3],[2, 4]])
        self.b3 = np.array([0.9, 0.2])

    def identity(self, x):
        return x
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def forward(self, x):
        # Forward pass through the network
        a1 = np.dot(x, self.w1) + self.b1
        z1 = self.sigmoid(a1)

        a2 = np.dot(z1, self.w2) + self.b2
        z2 = self.sigmoid(a2)

        a3 = np.dot(z2, self.w3) + self.b3
        y = self.identity(a3) # No activation function in output layer
        return y
```

```python
mlp = ThreeLayerMLP()
x = np.array([1.0, 0.5])
print(f'Input to the MLP = {x}')

y = mlp.forward(x)
print(f'Output to MLP = {y}')
```
